```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

# Assuming the dataset is stored in a CSV file named 'qsar_oral_toxicity.csv'
data = pd.read_csv('qsar_oral_toxicity.csv', header=None, sep=";")

data
```

```python
#data.columns()
```

```python
X = data.iloc[:, :-1]  # Features (binary molecular fingerprints)
y = data.iloc[:, -1]   # Target variable (experimental class)
y
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=100)
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(max_iter=1000)

# max_iter due to error message "error message you received is a ConvergenceWarning from scikit-learn's logistic regression implementation. "
```

```python
X_train
```

```python
y_train
```

```python
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

#lr = LogisticRegression(C=1)
#lr.fit(X_train, y_train)
#accuracy_score(y_test, lr.predict(X_test))
```

```python
#error_rate
print("test error rate: {:.2f}%".format((1-accuracy)*100))
```

in regards to the LDA/QDA application, it is not suitable to use on this dataset because it is a binary dataset. LDA/QDA method assumes that the predictor follows a Gaussian distribution as Gaussian distribution assumes the variables to be continuous that follow a bell-shaped curve. this is because Gaussian distribution is also referred as normal distribution

thus, in a binary dataset, the value is discrete, and usually it is categorised into 0 and 1. therefore, the discrete data contradicts the continous data required in Gaussian distribution. 

in addition, as Gaussian distribution is characterized by its mean, standard deviation, where majority of the data is concentrated around the mean, this Gaussian distribution that the value of the data can take any value within any real numbers in the range. unlike binary data, there are only two possible values and does not have a continuous range. therefore, it cannot fit adequately into the Gaussian distribution.


furthermore, LDA assumes that the variance of the predictors are equal across for each predictor variable. however it is not the case for this dataset.

In the case of binary molecular fingerprints, logistic regression is a more appropriate choice. Logistic regression can handle binary features and does not assume any particular distribution. It can capture the complex relationships between the binary molecular fingerprints and the experimental class without the need for distributional assumptions.
