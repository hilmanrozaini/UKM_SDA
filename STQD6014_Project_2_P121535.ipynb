## **EDA**
### Cleaning data : managing missing values

```python
 #to see the NA values dataset

LED.isna()
```

```python
 # sum of NA values are as below

LED.isna().sum()
```

```python
# to impute missing NA with mean

from sklearn.impute import SimpleImputer
imputer=SimpleImputer(missing_values=np.nan,strategy='median',fill_value=None)
LED['Life expectancy ']=imputer.fit_transform(LED[['Life expectancy ']])
LED['Adult Mortality']=imputer.fit_transform(LED[['Adult Mortality']])
LED['Alcohol']=imputer.fit_transform(LED[['Alcohol']])
LED['Hepatitis B']=imputer.fit_transform(LED[['Hepatitis B']])
LED[' BMI ']=imputer.fit_transform(LED[[' BMI ']])
LED['Polio']=imputer.fit_transform(LED[['Polio']])
LED['Total expenditure']=imputer.fit_transform(LED[['Total expenditure']])
LED['Diphtheria ']=imputer.fit_transform(LED[['Diphtheria ']])
LED['GDP']=imputer.fit_transform(LED[['GDP']])
LED['Population']=imputer.fit_transform(LED[['Population']])
LED[' thinness  1-19 years']=imputer.fit_transform(LED[[' thinness  1-19 years']])
LED[' thinness 5-9 years']=imputer.fit_transform(LED[[' thinness 5-9 years']])
LED['Income composition of resources']=imputer.fit_transform(LED[['Income composition of resources']])
LED['Schooling']=imputer.fit_transform(LED[['Schooling']])
```

```python
 # checking sum of NA values after mean imputation

LED.isna().sum()
```

```python
# describing the data
LED.describe()
```

```python
# Checking for duplicate records
LED.loc[LED.duplicated()]
```

## **Analysis**

### Life expectancy analysis

```python
plt.figure(figsize=(30,10))
ax = LED.groupby('Country')['Life expectancy '].agg('mean').plot(kind='bar', color='green')
ax.set_ylabel('Life expectancy (average)')
ax.set_xlabel('Country')
plt.show()
```

```python
# to observe which country is identified as developing or developed country
LED.loc[:,["Country","Status"]].drop_duplicates()
```

```python
f, ax = plt.subplots(figsize=(20,10))
ax = sns.lineplot(data=LED, x='Year', y='Life expectancy ', hue='Status')
```

It is observed that the average life expectancy for developing countries are approximately below 65 years in year 2000, and it is steadily increasing arpund 5 years to 70 years in 2014.

In contrast, the developed countries have more life expectancy then developing countries with approximately 77 years in year 2000, and it increasing as well to around 80 years in year 2014

Although both developing and developed countries showed an increased in of life expectancy over the years, it is to note that the average differences between the two status countries are roughly 10 years and that the differences has remained constant from year 2000 to year 2014.

### Hepatitis B immunization analysis

```python
plt.figure(figsize=(30,10))
ax1 = LED.groupby('Country')['Hepatitis B'].agg('mean').plot(kind='bar', color='green')
ax1.set_ylabel('Hepatitis B (average)')
ax1.set_xlabel('Country')
plt.show()
```

```python
f, ax1 = plt.subplots(figsize=(20,10))
ax1 = sns.lineplot(data=LED, x='Year', y='Hepatitis B', hue='Status')
```

This factor discusses the coverage for Hepatitis B immunization among 1 year old.

At initial, in year 2000, immunization coverage of Hepatitis B has higher coverage in developing countries than in developed countries. This can be assumed that the illness related to Hepatitis B is more critical in developing countries which requires a wider coverage. This can be further justified on the low life expectancy for developing countries as compared to developed countries.

Across the time frame, we can observe that the immunization coverage for developed countries is higher and remain constant except for year 2012.

As a result to the immunization doses for Hepatitis B, we can deduce that there is an increasing life expectancy gradient for both developing and developed countries.

However, the immunization coverage for developed countries has higher percentage than in developing countries. This may be related to the factors on government spending and income composition index where monetary resources can be scarce.

Next we will analyse the number of schooling as a social factor that may affect the life expectancy between developing and developed countries.

Therefore, the lower schooling years can be a worrying factor that affects the life expectancy of developing countries in the long run

### Income composition of resources analysis

```python
f, ax = plt.subplots(figsize=(20,10))
ax = sns.boxplot(data=LED, x='Year', y='Income composition of resources', hue='Status')
```

Human Development Index in terms of income composition of resources (index ranging from 0 to 1)

This factor discusses on Human Development Index in terms of income composition of resources where it focuses on the income distribution across the population.

From the boxplot, we can observe that the average median of income index for developed countries are around 0.8, and it is always higher than the developing countries with index reading of 0.6. Furthermore, the upper and lower quartile for all developed countries are also higher than developing countries, and we can deduce that at least 25% of the population in developed countries' income composition are always higher than the developing countries.

In regards to the immunization plot, the disparity of income between developed and developing countries can explain the immunization coverage as it is higher in developed countries. This may due to higher disposable income composition which allows them to afford the immunzation doses as compared to the population in developing countries.

Furthermore, we can observe the whiskers for all developing countries are wider than in developed countries. This means that the range between the lower observation and higher observation is large in developing countries.

In year 2000, we can observe where the lower income group can be as low as 0.0 index and more than 0.8 index for the higher income group. This disparity signals the potential causes for consistant differences in both number of schooling years and immunuzation coverage. This is because with higher income composition, more resources can be invested and disposed to address the health and educational needs in a population, to which is highly scarce in developing countries.

although the median and interquartile range for developing countries has increased from year 2000 to 2014, the spread between the boxplot fence is still consistently large as compared to the developed countries. Thus as a result, it indirectly affects the life expectancy of the average population in developing countries to be lower consistently.

Therefore, government assistance is needed to manage the large spread of income composition in developing countries to which we will observe on government expenditure in the next analysis.

### Total government expenditure on healthcare analysis

```python
f, ax = plt.subplots(figsize=(20,10))
ax = sns.boxplot(data=LED, x='Year', y='Total expenditure', hue='Status')
```

In this discussion, we will observe total government expenditure on healthcare against the countries' total expenditure in terms of percentage.

In overall, we observe that the developed countries expenditure is approximately 7.5% of the countries' total expenditure as opposed to developing countries at around 5% in total healthcare expenditure.

Government spending power is also depends on the population income composition as government fund is obtain through federal taxes. Since income composition is consistently lower in developing countries than in developed countries, it is justified that the government expenditure is also consistently lower.

However we can observe that there has been an increase in the total government expenditure over the years, which also led to an increase in life expectancy in both developed and developing countries. This shows an improvement and progressive growth towards improving the population well-being and healthcare.
